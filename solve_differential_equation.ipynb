```python
from gekko import GEKKO
import numpy as np
import matplotlib.pyplot as plt  
m = GEKKO(remote=False)
```

```python
# better than odeint inscipy cuz if k is parameter or multiple diff equations than gekko is better
# or stiff equations or DAE

k = 10
m.time = np.linspace(0,20,100) # without this it can't solve, time points

y = m.Var(value=5.0)
t = m.Param(value=m.time)
m.Equation(k*y.dt()==-t*y)
m.options.IMODE=4 # dynamic simulation
m.solve(disp=False)
```

```python
plt.plot(m.time,y.value)
plt.xlabel('time')
plt.ylabel('y')
plt.show()
```
